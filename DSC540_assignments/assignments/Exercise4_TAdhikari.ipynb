**1. Handling Missing Data**

We will check which columns have the most missing data. If there are columns with more than 60% of the data missing, we will manually delete those

```python
# check for missing value for Each column
def percentMissing(df):
    
    df_numeric = df.select_dtypes(include=[np.number])
    numeric_cols = df_numeric.columns.values
    print(numeric_cols)
    
    # % of missing data
    for col in df.columns:
        # create missing indicator for features with missing data
        missing = df[col].isnull()   # checking null value for each col
        pct_missing = np.mean(missing)*100
        print('{} - {}%'.format(col, round(pct_missing)))
```

```python
# check for missing value for Each column
def percentMissing_Col(df):
    
    df_numeric = df.select_dtypes(include=[np.number])
    numeric_cols = df_numeric.columns.values
    print(numeric_cols)
    
    # % of missing data
    for col in df.columns:
        # create missing indicator for features with missing data
        missing = df[col].isnull()   # checking null value for each col
        pct_missing = np.mean(missing)*100
        if pct_missing >=60:
            print('{} - {}%'.format(col, round(pct_missing)))
```

We will check which rows have the most missing data. If there are rows with more than 80% of the data missing, meaning if an observation is missing values from 91 columns out of 113, then we will remove those observations.

```python
# check for missing value for Each observation
def percentMissing_row(df):
    index_num = []
    # % of missing data
    for i in range(len(df.index)):
        # create missing indicator for features with missing data
        missing = df.iloc[i].isnull().sum()   # checking null value for each observation
        pct_missing = (missing/113)*100
        if pct_missing >=80:
            print('row index {} - {}%'.format(i, round(pct_missing)))
            index_num.append(i)
    return index_num
```

**3. String Manipulation**

```python
def no_whitespace(string):
    return string.strip()

# stripping white space from all column type as object
def clean_string(df):
    
    for col in df.columns:
        for i, row in df.iterrows():
            if df.at[i, col] == np.object:
                df.at[i, col] = no_whitespace(df.at[i, col])
    return df

df = clean_string(df)
```
