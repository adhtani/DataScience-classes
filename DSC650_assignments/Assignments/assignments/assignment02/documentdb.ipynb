```python
from pathlib import Path
import json
import os

from tinydb import TinyDB

current_dir = Path(os.getcwd()).absolute()
results_dir = current_dir.joinpath('results')
kv_data_dir = results_dir.joinpath('kvdb')
kv_data_dir.mkdir(parents=True, exist_ok=True)


class DocumentDB(object):
    def __init__(self, db_path):
        ## You can use the code from the previous exmaple if you would like
        people_json = kv_data_dir.joinpath('people.json')
        visited_json = kv_data_dir.joinpath('visited.json')
        sites_json = kv_data_dir.joinpath('sites.json')
        measurements_json = kv_data_dir.joinpath('measurements.json')
        
        self._sites = json.load(open(sites_json))
        self._visits = json.load(open(visited_json))
        self._persons = json.load(open(people_json))
        self._measurements = json.load(open(measurements_json))

        self._db_path = Path(db_path)
        self._db = None
        ## TODO: Implement code
        self._load_db()
        
        

    def _load_db(self):
        self._db = TinyDB(self._db_path, indent=4, separators=(',', ': '))
        sites = self._sites
        visits = self._visits
        persons = self._persons
        measurements = self._measurements
        
          
        for person in persons:
            x = persons[person]['person_id']
            x_measurement = self.get_measurements(x, measurements)
            x_visits = self.get_visits(x_measurement, visits, sites)
            persons[person]['visits'] = x_visits
            self._db.insert(persons[person])
        
     
    def get_measurements(self, x, measurements):
        x_measurement = []
        for i in measurements:
            if x in measurements[i]['person_id']:
                x_measurement.append(measurements[i])
        return x_measurement
            
            
    # returns list of visits by measurement for each person 
    
    def get_visits(self, x_measurement, visits, sites):
        uniqueVisitIds = []
        for m in x_measurement:
            visit_id = m['visit_id']
            if(visit_id not in uniqueVisitIds):
                uniqueVisitIds.append(visit_id)
        
        
        # returns sites for each visit
        def get_sites(x, sites):
            if x in sites.keys():
                return sites[x]
        allFoundVisits = []
        for vid in uniqueVisitIds:
            foundVisit = []
            for visit in visits:
                visit_id = visits[visit]['visit_id']
                if(visit_id == vid):
                    foundVisit.append(visits[visit])
                    #print(visits[visit])
                    break
            if len(foundVisit):
                #print(foundVisit[0]['site_id'])
                site_id = foundVisit[0]['site_id']
                foundVisit[0]['site'] = get_sites(site_id, sites)
                measurementList = []
                for m in x_measurement:
                    visit_id = m['visit_id']
                    if(foundVisit[0]['visit_id'] == visit_id):
                        measurementList.append(m)
                foundVisit[0]['measurement'] = measurementList
                allFoundVisits.append(foundVisit[0])
        return allFoundVisits
```

```python
db_path = results_dir.joinpath('patient-info.json')
if db_path.exists():
    os.remove(db_path)

db = DocumentDB(db_path)
```
