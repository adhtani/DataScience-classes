1. You can find the dataset controversial-comments.jsonl for this exercise in the Weekly Resources: Week 2 Data Files.

    Pre-processing Text: For this part, you will start by reading the controversial-comments.jsonl file into a DataFrame. Then,

    A. Convert all text to lowercase letters.

```python
df = pd.read_json('controversial-comments.jsonl', lines=True)
df.count()
```

```python
df.head(6)
```

```python
# randomly selecting the sample
sample = df.sample(n = 50000)
sample.count()
```

```python
sample.head(6)
```

```python
# converting all text to lowercase
sample["txt"] = sample["txt"].str.lower()
sample.head(3)
```

    B. Remove all punctuation from the text.

```python
# removing punctuation using string.punctuations and join()
# it joins each string if it's not a punctuation

import string
sample['txt'] = sample['txt'].apply(lambda x: "".join([i for i in x if i not in string.punctuation]))

sample.head(6)
```

    C. Remove stop words.

```python
import nltk
nltk.download('stopwords')
```

```python
# of NLTK stopwords in each row text
from nltk.corpus import stopwords
stop = stopwords.words('english')

sample['stopwords'] = sample['txt'].apply(lambda x: len([i for i in x.split() if i in stop]))
sample[['txt','stopwords']].head()
```

```python
# removing stop words from each row
sample['txt'] = sample['txt'].apply(lambda x: " ".join(i for i in x.split() if i not in stop))
sample['txt'].head(4)
```

```python
# check to see if it removed all stop words
sample['stopwords'] = sample['txt'].apply(lambda x: len([i for i in x.split() if i in stop]))
sample[['txt','stopwords']].head(4)
```

    D. Apply NLTK’s PorterStemmer.

```python
# Apply NLTK’s PorterStemmer
from nltk.stem import PorterStemmer
porter = PorterStemmer()
sample['txt'] = sample['txt'].apply(lambda x: " ".join([porter.stem(word) for word in x.split()]))
sample['txt'].head(6)
```

2. Now that the data is pre-processed, you will apply three different techniques to get it into a usable form for model-building. Apply each of the following steps (individually) to the pre-processed data.

    A. Convert each text entry into a word-count vector (see sections 5.3 & 6.8 in the Machine Learning with Python Cookbook).

For this part I am reducing the data because I am getting a MemoryError

```python
# randomly selecting the sample
data = sample.sample(n = 10000)
data.count()
data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# Create array with each text
data_arr = np.array(data['txt'].values.astype("U"))

# Create the bag of words feature matrix
counter = CountVectorizer(lowercase=True, ngram_range=(1,1),analyzer = "word")
bag_of_words = counter.fit_transform(data_arr)
arr = bag_of_words.toarray()
col = counter.get_feature_names()
```

```python
col
```

```python
pd.DataFrame(arr, columns = col)
```

    B. Convert each text entry into a part-of-speech tag vector (see section 6.7 in the Machine Learning with Python Cookbook).

```python
from nltk import pos_tag, word_tokenize
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
text_list = data['txt'].tolist()

tagged_texts = []

# tag each word
for text in text_list:
    text_tag = nltk.pos_tag(word_tokenize(text))
    tagged_texts.append([tag for word, tag in text_tag])
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

# one-hot encoding to convert the tags into features
one_hot_multi = MultiLabelBinarizer()
tags = one_hot_multi.fit_transform(tagged_texts)

col = one_hot_multi.classes_
```
