```python
import numpy as np
import pandas as pd
```

1. Create an array of size [10,50] and use the random function to fill it with integers between 1 and 500 (inclusive). Load it to a DataFrame.

```python
# array of random numbers
data = np.random.randint(1,501,size=(10,50))

# array into dataframe
df = pd.DataFrame(data)
df
```

2. Calculate the sum of each row; the sum of each column; and the sum of all entries.

```python
# sum of column
print("Sum of each column")
df.sum(axis=0)
```

```python
# sum of rows
print("Sum of each row")
df.sum(axis=1)
```

```python
# sum of all values
sum_values = np.sum(df.values)
print("Sum of all values = {}".format(sum_values))
```

3. Determine the minimum; maximum; average; and standard deviation of all entries.

```python
min_value = np.min(df.values)
max_value = np.max(df.values)
mean_value = np.mean(df.values)
std_value = np.std(df.values)

print("Minimum Value = {}".format(min_value))
print("Maximum Value = {}".format(max_value))
print("Average = {}".format(mean_value))
print("Standard Deviation = {}".format(std_value))
```

4. Sort the DataFrame: by rows, by columns, on row 2, on column 2.

```python
# sorted by row 2
df_sorted_by_row = df.sort_values(by=2, axis=0)
print("Sorted by row 2 values")
df_sorted_by_row
```

```python
# sorted by column 2
df_sorted_by_col = df.sort_values(by=2, axis=1)
print("Sorted by column 2 values")
df_sorted_by_col
```
